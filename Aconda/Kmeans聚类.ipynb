```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
import scipy as sp
%matplotlib inline
```

```python
result =[]
x=0
filepath = 'data_202106.tsv'
import codecs
with codecs.open(filepath,'rb','gb18030',errors='ignore') as csvfile:
    for line in csvfile:
        temp1=line.split('\t')
        result.append(temp1[5:7])
           
result.pop(0)
for i in range(len(result)):
    result[i][0]=float(result[i][0])
    round(result[i][0],1)
    result[i][1]=int(result[i][1])
```

```python
a=np.array(result)
from sklearn import preprocessing
a=preprocessing.scale(a)
```

```python
plt.figure(figsize=(12,12))
plt.scatter(a[:,0],a[:,1])
plt.title('label uknown')
```

```python
from sklearn.cluster import KMeans

estimator=KMeans(n_clusters=3)
estimator.fit(a)
label_pred=estimator.labels_
```
