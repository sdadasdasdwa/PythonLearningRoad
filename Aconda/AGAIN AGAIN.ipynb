```python
import pandas as pd
import numpy as np
import random
import  matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.cluster import KMeans
```

```python
#读文件
data=pd.read_csv('again.csv')
data['label']=0
```

```python
data['Label1'].value_counts()
```

```python
#树类型的定义
class TreeNode(object):
    def __init__(self, name):
        self.name = name
        self.parent = None
        self.child = []

    def __repr__(self) :
        return '%s' % self.name

    def __str__(self):
        return str(self.name)

    def add_child(self, name):
        for n in range(len(self.child)):
            if name == self.child[n].name:
                return self.child[n]
        obj = TreeNode(name)
        obj.parent = self
        self.child.append(obj)
        return obj

    def __len__(self):
        return len(self.child)

    def dump(self, indent=0):
        tab = '    '*(indent-1) + ' |- ' if indent > 0 else ''
        print('%s%s' % (tab, self.name))
        for i in range(len(self.child)):
            self.child[i].dump(indent+1)
            
    def func(self):
        y=[]
        for i in range(len(self.child)):
            for j in range(len(self.child[i].child)):
                for k in range(len(self.child[i].child[j].child)):
                    for l in range(len(self.child[i].child[j].child[k].child)):
                            x=[]
                            for m in range(len(self.child[i].child[j].child[k].child[l])):
                                x.append(self.child[i].child[j].child[k].child[l].child[m])
                            y.append(x)
            return y
```

```python
R0=TreeNode(0)
n=data[data['Label1']==0]
for j in range(n.shape[0]):
    a=R0.add_child(n.iloc[j,11])
    b=a.add_child(n.iloc[j,12])
    c=b.add_child(n.iloc[j,13])
    d=c.add_child(n.iloc[j,14])
    d.add_child(n.iloc[j,0])
```

```python
L0=R0.func()
for i in range(len(L0)):
    for j in range(len(L0[i])):
        L0[i][j]=L0[i][j].name
        
d=[[]]
for i in range(len(L0)):
    if(len(L0[i])<1000):
        d[0].extend(L0[i])
    else:
        d.append(L0[i])
if(len(d[0])==0):
    d.pop(0)

for i in range(len(L0)):
    if(len(L0[i])>10000):
        a=random.sample(L0[0],round(len(L0[0])/10))
    elif(len(L0[i])>5000):
        a=random.sample(L0[0],round(len(L0[0])/5))
    else:
        a=L0[i]
    result=[]
    for i in range(len(a)):
        result.append([data.iloc[a[i],3],data.iloc[a[i],4]])
    train=np.array(result)
    train=preprocessing.scale(train)
    estimator=KMeans(n_clusters=7)
    estimator.fit(train)
    label_pred=estimator.labels_.copy()
    center=estimator.cluster_centers_.tolist()
    xy=center.copy()
    for i1 in range(7):
        for i2 in range(2):
            center[i1][i2]=round(center[i1][i2],2)
        center[i1]=round(center[i1][0]**2+center[i1][1]**2,4)
    center_ordinary=center.copy()
    center.sort()
    label_pred[label_pred==center_ordinary.index(center[0])]=7
    label_pred[label_pred==center_ordinary.index(center[1])]=7
    m=center_ordinary.index(center[-1])
    n=center_ordinary.index(center[-2])
    if(xy[m][0]>xy[n][0] and xy[m][1]<xy[n][1]):
        label_pred[label_pred==m]=8
        label_pred[label_pred==n]=9
    elif(xy[m][0]<xy[n][0] and xy[m][1]>xy[n][1]):
        label_pred[label_pred==m]=9
        label_pred[label_pred==n]=8
    elif(xy[m][0]<xy[m][1] and xy[n][0]<xy[n][1]):
        label_pred[label_pred==m]=9
        label_pred[label_pred==n]=9    
    else:
        label_pred[label_pred==m]=8
        label_pred[label_pred==n]=8
    for i3 in range(len(a)):
        if(label_pred[i3]==7  or label_pred[i3]==8 or label_pred[i3]==9):
            data.iloc[a[i3],15]=label_pred[i3]
```
